### Combinatorio recursivo

Codigo simple del combinatorio visto en clase:

```python
def combinatorio(n, k):
    if k == 0 or k == n:
        return 1
    else:
        a = combinatorio(n - 1, k - 1)
        b = combinatorio(n - 1, k)
        return a + b
    
# Ejemplo de uso
n = 5
k = 2
print(f"El número combinatorio C({n}, {k}) es: {combinatorio(n, k)}")
```

Veamos cuantas llamadas hace

```python
total_llamadas = 0

def combinatorio(n, k):
    global total_llamadas
    total_llamadas += 1
    print(f"combinatorio({n}, {k}) llamado")
    if k == 0 or k == n:
        return 1
    else:
        a = combinatorio(n - 1, k - 1)
        b = combinatorio(n - 1, k)
        return a + b
    
# Ejemplo de uso
n = 15
k = 5
print(f"El número combinatorio C({n}, {k}) es: {combinatorio(n, k)}")
print(f"Total de llamadas realizadas: {total_llamadas}")
```

Si ya llamé varias veces al mismo combinatorio, no es mejor guardarlo en una tabla?

Memoización en combinatorio top-down:

```python
total_llamadas = 0

def combinatorio_memo(n, k, memo={}):
    global total_llamadas
    total_llamadas += 1
    if (n, k) in memo:
        return memo[(n, k)]
    if k == 0 or k == n:
        return 1
    else:
        a = combinatorio_memo(n - 1, k - 1, memo)
        b = combinatorio_memo(n - 1, k, memo)
        memo[(n, k)] = a + b
        return memo[(n, k)]

# Ejemplo de uso
n = 15
k = 5
memo = {}
print(f"El número combinatorio C({n}, {k}) es: {combinatorio_memo(n, k, memo)}")
print(f"Total de llamadas realizadas: {total_llamadas}")

print("Tabla de memoización:")
for key in sorted(memo.keys()):
    print(f"C{key} = {memo[key]}")
```

Enfoque bottom up guardando los resultados en una tabla completa C.
Esto se puede optimizar para solo guardar la última fila!

```python
def combinatorio_bottom_up(n, k):
    C = [[0 for _ in range(k + 1)] for _ in range(n + 1)]
    
    for i in range(n + 1):
        C[i][0] = 1
    
    for j in range(1, k + 1):
        C[j][j] = 1
    
    for i in range(n + 1):
        for j in range(min(i, k) + 1):
            if j != 0 and j != i:
                C[i][j] = C[i - 1][j - 1] + C[i - 1][j]
                    
    return C[n][k]

# Ejemplo de uso
n = 5
k = 3
print(f"El número combinatorio C({n}, {k}) es: {combinatorio_bottom_up(n, k)}")
```

### Fibonacci

Recursivo:

```python
def fib(n, memo = {}):

    print(f"Fib({n})")
    if n <= 1:
        return n
    else:
        return fib(n - 1) + fib(n - 2)
    
a = 5
fib(a)
```

```python
def fib(n, memo = {}):
    if n in memo.keys():
        return memo[n]
    print(f"Fib({n})")
    if n <= 1:
        memo[n] = n
        return n
    else:
        memo[n] = fib(n - 1) + fib(n - 2)
        return memo[n]
    
a = 5
fib(a)
```

### El problema del cambio

Recursivo:

```python
llamados = 0
def monedas(s, denominaciones):    
    global llamados
    llamados+=1
    # caso base
    if s == 0:
        return 0
    
    pruebas = []
    
    for i in range(0, len(denominaciones)):
        if denominaciones[i] <= s:
            pruebas.append(1 + monedas(s - denominaciones[i], denominaciones))
    
    mejor = min(pruebas)

    return mejor


vuelto = 10
denominaciones = [50, 25, 10, 5, 1]

cantidad = monedas(vuelto, denominaciones)

print("llamados", llamados, "cantidad", cantidad)
```

Top-down con memoización:

```python
llamados = 0
def monedas(s, denominaciones, memo = {}):
    if s in memo.keys():
        return memo[s]
    
    global llamados
    llamados+=1
    
    # caso base
    if s == 0:
        return 0
    
    pruebas = []
    
    for i in range(0, len(denominaciones)):
        if denominaciones[i] <= s:
            pruebas.append(1 + monedas(s - denominaciones[i], denominaciones, memo))
    
    mejor = min(pruebas)
    
    memo[s] = mejor
    print(s, mejor)
    return mejor


vuelto = 69
denominaciones = [50, 25, 10, 5, 1]

cantidad = monedas(vuelto, denominaciones, {})

print("llamados", llamados, "cantidad", cantidad)
```

Bottom-up devolviendo que monedas se usan:

```python
import math
infinito = math.inf
import numpy as np

def monedas_botup(vuelto):
    memo = [0] * (vuelto + 1)
    que_monedas = [[]] * (vuelto + 1)
    denominaciones = [1, 5, 10, 25, 50]

    for s in range(1, vuelto+1):
        pruebas = [infinito] * len(denominaciones)
        for i in range(0, len(denominaciones)):
            if denominaciones[i] <= s:
                pruebas[i] = 1 + memo[s - denominaciones[i]]
        
        memo[s] = min(pruebas)
        
        optima = np.argmin(pruebas)
        que_monedas[s] = que_monedas[s - denominaciones[optima]] + [denominaciones[optima]]
    
    for i in range(0, s):
        print(i, que_monedas[i])
    
    return memo[vuelto], que_monedas[vuelto]

vuelto = 69
monedas_botup(vuelto)
```
